```python
# Convert date column to datetime format (instead of object)
df['date'] = pd.to_datetime(df['date'])

# set date as index because we have 10min intervals i.e uniqueness 
df.set_index('date', inplace=True) 
```

```python
df.describe() # summary statistics of numerical columns
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Appliances'], bins=50, kde=True)
plt.title("Distribution of Appliance Energy Consumption")
plt.xlabel("Energy Consumption (kWh)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Temporal tendency 
plt.figure(figsize=(12, 6))
df['Appliances'].resample('D').mean().plot()
plt.title('Daily Average Energy Consumption of Appliances')
plt.xlabel('Date')
plt.ylabel('Energy Consumption (kWh)')
plt.show()
```

```python
# To analyse temporal tendencies and have meaningful features, we create the hour, day of the week, month and year features from the date index
df['hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek
df['month'] = df.index.month
```

```python

# Average energy consumption per hour, day of the week, and month
hourly_consumption = df.groupby('hour')['Appliances'].mean()
weekly_consumption = df.groupby('day_of_week')['Appliances'].mean()
monthly_consumption = df.groupby('month')['Appliances'].mean()

fig, axes = plt.subplots(3, 1, figsize=(12, 12))

axes[0].plot(hourly_consumption.index, hourly_consumption.values, marker='o', linestyle='-', color='b')
axes[0].set_title("Consommation Moyenne d'Énergie par Heure")
axes[0].set_xlabel("Heure")
axes[0].set_ylabel("Consommation (Wh)")

axes[1].bar(weekly_consumption.index, weekly_consumption.values, color='g', alpha=0.7)
axes[1].set_title("Consommation Moyenne d'Énergie par Jour de la Semaine")
axes[1].set_ylabel("Consommation (Wh)")

axes[2].plot(monthly_consumption.index, monthly_consumption.values, marker='s', linestyle='-', color='r')
axes[2].set_title("Consommation Moyenne d'Énergie par Mois")
axes[2].set_xlabel("Mois")
axes[2].set_ylabel("Consommation (Wh)")

plt.tight_layout()
plt.show()
```

```python

# Temperature in the house 
temperature_cols = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'T9', 'T_out']
correlation_temp = df[temperature_cols + ['Appliances']].corr()['Appliances'].sort_values(ascending=False)

# Humidity in the house 
humidity_cols = ['RH_1', 'RH_2', 'RH_3', 'RH_4', 'RH_5', 'RH_6', 'RH_7', 'RH_8', 'RH_9', 'RH_out']
correlation_humidity = df[humidity_cols + ['Appliances']].corr()['Appliances'].sort_values(ascending=False)

# Weather conditions (external)
weather_cols = ['T_out', 'Press_mm_hg', 'RH_out', 'Windspeed', 'Visibility', 'Tdewpoint']
correlation_weather = df[weather_cols + ['Appliances']].corr()['Appliances'].sort_values(ascending=False)

print(correlation_temp)
print(correlation_humidity)
print(correlation_weather)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 15))

# Outdoor Temperature vs Energy Consumption
sns.scatterplot(x=df['T_out'], y=df['Appliances'], alpha=0.5, ax=axes[0])
axes[0].set_title("Outdoor Temperature vs Energy Consumption")
axes[0].set_xlabel("Outdoor Temperature (°C)")
axes[0].set_ylabel("Energy Consumption (Wh)")

# Outdoor Humidity vs Energy Consumption
sns.scatterplot(x=df['RH_out'], y=df['Appliances'], alpha=0.5, ax=axes[1])
axes[1].set_title("Outdoor Humidity vs Energy Consumption")
axes[1].set_xlabel("Outdoor Humidity (%)")
axes[1].set_ylabel("Energy Consumption (Wh)")

# Wind Speed vs Energy Consumption
sns.scatterplot(x=df['Windspeed'], y=df['Appliances'], alpha=0.5, ax=axes[2])
axes[2].set_title("Wind Speed vs Energy Consumption")
axes[2].set_xlabel("Wind Speed (m/s)")
axes[2].set_ylabel("Energy Consumption (Wh)")

plt.tight_layout()
plt.show()
```

```python
# Impact of light on energy consumption
correlation_lights = df[['lights', 'Appliances']].corr()['Appliances'].sort_values(ascending=False)
print(correlation_lights)
# Scatter plot of lights vs energy consumption
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['lights'], y=df['Appliances'], alpha=0.5)
plt.title("Impact of Lights on Energy Consumption")
plt.xlabel("Lights (Wh)")
plt.ylabel("Energy Consumption (Wh)")
plt.show()
```
